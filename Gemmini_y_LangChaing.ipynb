Instalación de herramientas esenciales: Instalar Langchain y sus complementos para OpenAI es un paso fundamental para la integración

```python
import os
from google.colab import userdata
from google import genai

# Obtiene la clave del Secret de Colab
GOOGLE_API_KEY = userdata.get('GEMINI_API_KEY')

# Configura el cliente de Gemini
client = genai.Client(api_key=GOOGLE_API_KEY)

# Ejemplo de uso
response = client.models.generate_content(
    model="gemini-2.5-flash",
    contents="Explica cómo funciona la IA en pocas palabras."
)

print(response.text)
```
